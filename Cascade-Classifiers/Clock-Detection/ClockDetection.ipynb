```python
image = cv.imread('clock.jpg')
cv2_imshow(image)
```

```python
clock_detector = cv.CascadeClassifier('/clocks.xml')
clock_detection = clock_detector.detectMultiScale(image, scaleFactor=1.025)
for (x, y, w, h) in clock_detection:
  cv.rectangle(image, (x, y), (x + w, y + h), (0, 255, 0), 2)
cv2_imshow(image)
```

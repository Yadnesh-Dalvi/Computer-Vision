```python
import cv2
```

```python
image = cv2.imread('/content/People.jpg')
```

```python
image.shape
```

```python
from google.colab.patches import cv2_imshow
cv2_imshow(image)
```

```python
image = cv2.resize(image, (800,600))
cv2_imshow(image)
```

Converting image into Grey Scale image

```python
image_grey = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2_imshow(image_grey)
```

```python
image.shape
```

```python
image_grey.shape
```

Detecting Face

```python
face_detector = cv2.CascadeClassifier('/content/drive/MyDrive/Udemy/Cascades/haarcascade_frontalface_default.xml')
```

```python
detections = face_detector.detectMultiScale(image_grey, scaleFactor = 1.3)
```

```python
detections
```

```python
len(detections)
```

```python
for (x,y,w,h) in detections:
  cv2.rectangle(image, (x,y), (x+w, y+h), (0,255,0), 2)
cv2_imshow(image)
```

Eye Detection

```python
eye_detector = cv2.CascadeClassifier('/content/drive/MyDrive/Udemy/Cascades/haarcascade_eye.xml')
eye_detections = eye_detector.detectMultiScale(image_grey, scaleFactor = 1.3)
for (x,y,w,h) in eye_detections:
  cv2.rectangle(image, (x,y), (x+w, y+h), (255,0,0), 2)
cv2_imshow(image)
```
